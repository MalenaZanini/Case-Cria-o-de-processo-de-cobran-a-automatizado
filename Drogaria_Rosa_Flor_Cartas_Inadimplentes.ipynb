```python
# importando bibliotecas necessarias

from docx import Document
import pandas as pd
import re
from docx.shared import Inches, Pt
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

```

```python
# lendo a planilha 

planilha = pd.read_excel("/content/drive/MyDrive/Drogaria Rosa Flor/Pessoas em espera 2.xlsx")
planilha.head(20)
```

```python
# Para não ter que repetir o mesmo comando para cada cliente, vamos criar uma função chamada "criar_carta_unica" e ter como parametro a base com os dados dos clientes

def criar_carta_unica(base):
    doc = Document()

    # Primeiramente ajustar as margens do documento

    section = doc.sections[0]
    section.top_margin = Inches(0.3)
    section.bottom_margin = Inches(0.3)
    section.left_margin = Inches(0.3)
    section.right_margin = Inches(0.3)

    # Estilo do documento
    style = doc.styles['Normal']
    style.font.name = 'Calibri'
    style.font.size = Pt(9.7)

    # Loop para percorrer a planilha e gerar cartas
    for index, row in base.iterrows():
        nome = row['Cliente']
        valor = row['Valor Aberto']

        # Adicionando nome da farmácia em negrito e tamanho 14

        titulo = doc.add_paragraph()
        run = titulo.add_run("Drogaria Rosa Flor")
        run.bold = True
        run.font.size = Pt(14)
        titulo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT
        titulo.paragraph_format.space_after = Pt(0)
        titulo.paragraph_format.space_before = Pt(0)

        # Endereço e telefone
        paragrafo_endereco = doc.add_paragraph("R. Quinze de Novembro - Centro, Beneditinos - PI, 64380-000")
        paragrafo_endereco.paragraph_format.space_before = Pt(0)
        paragrafo_endereco.paragraph_format.space_after = Pt(0)

        # Saudação com adaptação para o gênero
        paragrafo = doc.add_paragraph("Prezado cliente, ")
        run = paragrafo.add_run(nome)
        run.bold = True
        paragrafo.add_run(". Esperamos que esteja bem.")
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT
        paragrafo.paragraph_format.space_after = Pt(0)
        paragrafo.paragraph_format.space_before = Pt(0)

        # Corpo da carta com o valor formatado
        valor_formatado = f"R$ {valor:.2f}".replace('.', ',')

        paragrafo = doc.add_paragraph(
        "Estamos entrando em contato para informar que identificamos pendências financeiras em sua conta conosco, que totalizam o valor de  "
        )

        # Adicionando o valor em negrito no mesmo parágrafo
        run = paragrafo.add_run(valor_formatado)
        run.bold = True  # Tornando o valor em negrito
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT
        paragrafo.paragraph_format.space_after = Pt(0)
        paragrafo.paragraph_format.space_before = Pt(0)

        paragrafo = doc.add_paragraph(
            "Compreendemos que imprevistos podem acontecer e que circunstâncias financeiras podem ser desafiadoras. "
            "Nosso objetivo é oferecer suporte e buscar, juntos, uma solução para regularizar esta pendência "
            "de forma que respeite suas condições atuais."
        )
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT  # Justificando o texto
        paragrafo.paragraph_format.space_after = Pt(0)  # Ajustando espaçamento após o parágrafo
        paragrafo.paragraph_format.space_before = Pt(0)  # Ajustando espaçamento antes do parágrafo
        paragrafo = doc.add_paragraph(
            "No entanto, gostaríamos de alertar que caso não seja realizada a regularização dessa pendência, "
            "o seu nome poderá ser incluído no cadastro de inadimplentes do Serasa. Sabemos que essa situação pode trazer "
            "dificuldades adicionais e queremos evitar que isso aconteça, oferecendo nossa colaboração para encontrar "
            "a melhor solução."
        )
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT  # Justificando o texto
        paragrafo.paragraph_format.space_after = Pt(0)  # Ajustando espaçamento após o parágrafo
        paragrafo.paragraph_format.space_before = Pt(0)  # Ajustando espaçamento antes do parágrafo
        paragrafo = doc.add_paragraph(
            "Gostaríamos de agendar uma conversa para explorar alternativas que melhor se adequem às suas necessidades. "
            "Estamos à disposição para negociar condições de pagamento ou esclarecer quaisquer dúvidas que possam surgir."
        )
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT  # Justificando o texto
        paragrafo.paragraph_format.space_after = Pt(0)  # Ajustando espaçamento após o parágrafo
        paragrafo.paragraph_format.space_before = Pt(0)  # Ajustando espaçamento antes do parágrafo
        # Destacando "em até 10 dias"
        paragrafo = doc.add_paragraph("Por favor, entre em contato conosco ")
        run = paragrafo.add_run("em até 10 dias")
        run.bold = True
        paragrafo.add_run(" pelo telefone (86) 99449-9491 ou compareça ao nosso estabelecimento.")
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT  # Justificando o texto
        paragrafo.paragraph_format.space_after = Pt(0)  # Ajustando espaçamento após o parágrafo
        paragrafo.paragraph_format.space_before = Pt(0)  # Ajustando espaçamento antes do parágrafo

        paragrafo = doc.add_paragraph(
            "Estamos confiantes de que, por meio do diálogo, encontraremos a melhor forma de resolver esta situação. "
            "Agradecemos pela sua atenção e pela oportunidade de continuar atendendo."
        )
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT  # Justificando o texto
        paragrafo.paragraph_format.space_after = Pt(0)  # Ajustando espaçamento após o parágrafo
        paragrafo.paragraph_format.space_before = Pt(0)  # Ajustando espaçamento antes do parágrafo

        run = paragrafo.add_run(" Caso este débito já tenha sido quitado, desconsidere esta carta, atenciosamente, "
        )
        run.bold = True  # Aplica negrito a toda a frase
        paragrafo.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT  # Justificando o texto
        paragrafo.paragraph_format.space_after = Pt(0)  # Ajustando espaçamento após o parágrafo
        paragrafo.paragraph_format.space_before = Pt(0)  # Ajustando espaçamento antes do parágrafo

        # Nome final em negrito
        paragrafo = doc.add_paragraph()
        run = paragrafo.add_run("Rosa Maria Braga.")
        run.bold = True

    return doc
```

```python
# Gerando o documento único com todas as cartas
doc = criar_carta_unica(planilha)

# Salvando o documento único no seu computador (na pasta do projeto)
arquivo_nome = "carta_de_cobranca_unica.docx"
doc.save(arquivo_nome)

print(f"Documento único salvo como '{arquivo_nome}'")
```
